## Import necessary Python packages and Load the Data

```python
import numpy as np;
import xarray as xr;
import matplotlib.pyplot as plt;

import cartopy.crs as ccrs;
import cartopy.mpl.ticker as cticker;
from cartopy.util import add_cyclic_point;

from scipy.stats import ttest_ind;

from scipy.stats import pearsonr

from scipy.stats import linregress as lr;

import warnings

from dask.diagnostics import ProgressBar
```

```python
warnings.filterwarnings( action = 'ignore' );
```

```python
## Read dataset into notebook.  
##Calculate the annual means simultaneously to remove the dominant effect of the seasonal cycle. This also decreases the amount of data in memory.  
sst_path = '/shared/cmip5/data/piControl/ocean/mon/Omon/tos/NASA-GISS.GISS-E2-H/r1i1p1/';
atmos_path = '/shared/cmip5/data/piControl/atmos/mon/Amon/';
ts_path = atmos_path + '/ts/NASA-GISS.GISS-E2-H/r1i1p1/';
pr_path = atmos_path + '/pr/NASA-GISS.GISS-E2-H/r1i1p1/';
sst_file = 'tos_Omon_GISS-E2-H_piControl_r1i1p1_2*.nc';
ts_file = 'ts_Amon_GISS-E2-H_piControl_r1i1p1_2*.nc';
pr_file =  'pr_Amon_GISS-E2-H_piControl_r1i1p1_2*.nc';
sst_ds = xr.open_mfdataset( ( sst_path + sst_file ), concat_dim = 'time' ).groupby( 'time.year' ).mean( dim = 'time' );
ts_ds = xr.open_mfdataset( ( ts_path + ts_file ), concat_dim = 'time' ).groupby( 'time.year' ).mean( dim = 'time' );
pr_ds = xr.open_mfdataset( ( pr_path + pr_file ), concat_dim = 'time' ).groupby( 'time.year' ).mean( dim = 'time' );
##Convert precipitation from mm/s to mm/day
pr_ds['pr'] = pr_ds['pr'] * 8.64 * ( 10 ** 4 );
sst_ds
## Note that there is no units attribute for this file.
```

```python
##Shift the longitude of the ocean dataset to be consistent with the atmospheric datasets
new_rlon = ( sst_ds['lon'] + 360 ) % 360;
sst_ds = sst_ds.assign_coords( lon = new_rlon );
sst_ds = sst_ds.sortby( sst_ds['lon'] );
sst_ds
```

## Calculate an AMV Index, from -80-0W and 0-60N

```python
sst_amv = sst_ds.sel( lat = slice( 0, 60 ),
                         lon = slice( 360-80, 360 ) );
sst_amv
```

```python
##Make the figure larger
fig = plt.figure( figsize = ( 11, 8.5 ) );

#Set the axes using the specified map projection
ax = plt.axes( projection = ccrs.PlateCarree() );

#Make a filled contour plot
cs = ax.contourf( sst_amv['lon'],
                 sst_amv['lat'],
                 sst_amv['tos'][0,:,:,],
                 transform = ccrs.PlateCarree(),
                 cmap = 'coolwarm', extend = 'both' );

#Add coastlines
ax.coastlines();

#Add gridlines
ax.gridlines();

##Define the xticks for longitude
ax.set_xticks( np.arange( -180, 181, 60 ), crs = ccrs.PlateCarree() );
lon_formatter = cticker.LongitudeFormatter();
ax.xaxis.set_major_formatter( lon_formatter );

##Define the yticks for latitude
ax.set_yticks( np.arange( -90, 91, 30 ), crs = ccrs.PlateCarree() );
lat_formatter = cticker.LatitudeFormatter();
ax.yaxis.set_major_formatter( lat_formatter );

#Add colorbar
cbar = plt.colorbar( cs, shrink = 0.7,
                    orientation = 'horizontal',
                    label = 'SST degK' );

#Add title.  Note that using .groupby().mean() removes the attributes that describe the dataset
plt.title( 'AMV Index Region, displaying first time step' );
```

```python
amv_ind = sst_amv['tos'].mean( dim = ( 'lat', 'lon' ) );
print( amv_ind );
plt.plot( amv_ind['year'], amv_ind );
plt.title( 'NASA GISS AMV Index Calculated via Area Averaging' );
```

### Center the Time Series

```python
amv_mean = amv_ind.mean();
amv_ind = amv_ind - amv_mean;
plt.plot( amv_ind['year'], amv_ind );
plt.title( 'AMV Index Anomalies' );
plt.xlabel( 'PiControl Year' );
plt.ylabel( 'AMV Anomaly degC' )
```

## Determine the Composite maps of Precipitation and Surface Temperature relative to the calculated AMO index

Based on a visual inspection of the variability of the Area Averaged NASST Index, thresholds are chosen at +- 0.2 degrees celsius

```python
##By omitting return values for the where() function, Nans will be returned where the condition is not met
pos_amv = amv_ind.where( amv_ind >= 0.1 );
neg_amv = amv_ind.where( amv_ind <= -0.1 );
neutral = amv_ind.where( np.logical_and( amv_ind > -0.1, amv_ind < 0.1 ) );
##The 'and' command will attempt to find a single true/false value
```

```python
print( 'Positive AMV: ', pos_amv.count( dim = 'year' ).values );
print( 'Negative AMV: ', neg_amv.count( dim = 'year' ).values );
print( 'Neutral: ', neutral.count( dim = 'year' ).values );

counts = [pos_amv.count( dim = 'year' ).values, 
         neg_amv.count( dim = 'year' ).values, 
         neutral.count( dim = 'year' ).values ];
print( counts );
```

```python
plt.plot( amv_ind['year'][0:50], amv_ind[0:50], 'k' );
plt.fill_between( amv_ind['year'][0:50], 0.1, pos_amv[0:50], color = 'red' );
plt.fill_between( amv_ind['year'][0:50], neg_amv[0:50], -0.1, color = 'blue' );
plt.axhline( 0.1, color = 'black', linewidth = 0.1, linestyle = 'dotted' );
plt.axhline( -0.1, color = 'black', linewidth = 0.1, linestyle = 'dotted' );
plt.title( 'NASA GISS AMV Index Anomalies Displayed with Composite Limits' );
plt.ylabel( 'AMV Anomaly degC' );
plt.xlabel( 'PiControl Year' );
```

```python
plt.plot( amv_ind['year'], amv_ind, 'k' );
plt.fill_between( amv_ind['year'], 0.1, pos_amv, color = 'red' );
plt.fill_between( amv_ind['year'], neg_amv, -0.1, color = 'blue' );
plt.axhline( 0.1, color = 'black', linewidth = 0.1, linestyle = 'dotted' );
plt.axhline( -0.1, color = 'black', linewidth = 0.1, linestyle = 'dotted' );
plt.title( 'NASA GISS AMV Index Anomalies Displayed with Composite Limits' );
plt.ylabel( 'AMV Anomaly degC' );
plt.xlabel( 'PiControl Year' );
```

## Sea Surface Temperature

```python
sst_climo = sst_ds.mean( dim = 'year' );
sst_anoms = sst_ds - sst_climo;
sst_anoms
```

```python
pos_sst = sst_anoms.sel( year = pos_amv.dropna( dim = 'year' )['year'] ).mean( dim = 'year' );
neg_sst = sst_anoms.sel( year = neg_amv.dropna( dim = 'year' )['year'] ).mean( dim = 'year' );
neutral_sst = sst_anoms.sel( year = neutral.dropna( dim = 'year' )['year'] ).mean( dim = 'year' );

comp_sst = [pos_sst, neg_sst, neutral_sst];
comp_sst                      
```

```python
labels = ['Positive AMV', 'Negative AMV', 'Neutral AMV'];
clevs = np.arange( -1.4, 1.8, 0.4 );

## Define the figure and each axis for the 3 rows and 3 columns
fig, axs = plt.subplots( nrows = 3, ncols = 1,
                       subplot_kw = {'projection': ccrs.PlateCarree() },
                       figsize = (8.5, 11) );
## axs is a 2 dimensional array of 'GeoAxes'.
##We will flatten this into a 1D array
axs = axs.flatten();

##Loop over each of the conditions and plot
for i, nasst in enumerate( comp_sst ):
    ##Select the comp
    data = comp_sst[i]['tos'];
    ##Add the cyclic point
    data,lons = add_cyclic_point( data, coord = comp_sst[i]['lon'] );
    ##Create the contour plot
    cs = axs[i].contourf( lons, comp_sst[i]['lat'], data, clevs,
                        transform = ccrs.PlateCarree(),
                        cmap = 'RdBu_r', extend = 'both' );
    ##Longitude labels 
    axs[i].set_xticks( np.arange( -180, 181, 60 ), crs = ccrs.PlateCarree() );
    lon_formatter = cticker.LongitudeFormatter();
    axs[i].xaxis.set_major_formatter( lon_formatter );
    ##Latitude labels
    axs[i].set_yticks( np.arange( -90, 91, 30 ), crs = ccrs.PlateCarree() );
    lat_formatter = cticker.LatitudeFormatter();
    axs[i].yaxis.set_major_formatter( lat_formatter );
    ##Title each subplot with the name of the composite
    axs[i].set_title( 'SST Anomaly for ' + labels[i] + ' (' + str(counts[i]) + ') ' );
    ##Draw the coastlines for each subplot
    axs[i].coastlines();
    axs[i].gridlines();
    
    
##Adjust the location of the subplots to make room for colorbar
fig.subplots_adjust( bottom = 0.15, top = 0.9, left = 0.05, right = 0.95,
                   wspace = 0.1, hspace = 0.5 );
##Add a colorbar axis
cbar_ax = fig.add_axes( [0.2, 0.1, 0.6, 0.02] );
cbar = fig.colorbar( cs, cax = cbar_ax, orientation = 'horizontal', label = 'degC' );

plt.suptitle( 'NASA GISS Composite SST Anomalies during AMV Variations')
```

```python
ptest = 0.05; 

a = sst_anoms.sel( year = pos_amv.dropna( dim = 'year' )['year'] );
b = sst_anoms.sel( year = neutral.dropna( dim = 'year' )['year'] );
c = sst_anoms.sel( year = neg_amv.dropna( dim = 'year' )['year'] );

pos_tstat,pos_pval = ttest_ind( a['tos'], b['tos'], equal_var = False );
neg_tstat,neg_pval = ttest_ind( c['tos'], b['tos'], equal_var = False );
```

```python
pos_mask_diff = pos_sst['tos'].where( pos_pval < ptest );
neg_mask_diff = neg_sst['tos'].where( neg_pval < ptest );
```

```python
labels = ['Positive - Neutral', 'Negative - Neutral'];
clevs = np.arange( -1.4, 1.8, 0.4 );

fig, axs = plt.subplots( nrows = 2, ncols = 1,
                       subplot_kw = {'projection':ccrs.PlateCarree()},
                       figsize = (8.5,11) );
axs = axs.flatten();

#for i,enso in enumerate( diff_precip ):
data = pos_sst['tos'] - neutral_sst['tos'];

data,lons = add_cyclic_point( data, coord = sst_anoms['lon'] );
mask_data,lons = add_cyclic_point( pos_mask_diff, coord=sst_anoms['lon'] );

cs = axs[0].contourf( lons, pos_sst['lat'], data, clevs,
                        transform = ccrs.PlateCarree(),
                        cmap = 'RdBu_r', extend = 'both' );
axs[0].contourf( lons, sst_anoms['lat'], mask_data,
            [0,1], transform = ccrs.PlateCarree(), colors = 'None',
            hatches = ['.'], extend = 'both', alpha = 0 );
axs[0].set_title( labels[0] + ' Composite Difference of SST sorted by AMV' );
axs[0].coastlines();
axs[0].gridlines();
##Longitude Labels
axs[0].set_xticks( np.arange( -180, 181, 60 ), crs = ccrs.PlateCarree() );
lon_formatter = cticker.LongitudeFormatter();
axs[0].xaxis.set_major_formatter( lon_formatter );
##Latitude labels
axs[0].set_yticks( np.arange( -90, 91, 30 ), crs = ccrs.PlateCarree() );
lat_formatter = cticker.LatitudeFormatter();
axs[0].yaxis.set_major_formatter( lat_formatter );


data = neg_sst['tos'] - neutral_sst['tos'];

data,lons = add_cyclic_point( data, coord = sst_anoms['lon'] );
mask_data,lons = add_cyclic_point( neg_mask_diff, coord=sst_anoms['lon'] );

cs = axs[1].contourf( lons, neg_sst['lat'], data, clevs,
                        transform = ccrs.PlateCarree(),
                        cmap = 'RdBu_r', extend = 'both' );
axs[1].contourf( lons, sst_anoms['lat'], mask_data,
            [0,1], transform = ccrs.PlateCarree(), colors = 'None',
            hatches = ['.'], extend = 'both', alpha = 0 );
axs[1].set_title( labels[1] + ' Composite Difference of SST sorted by AMV' );
axs[1].coastlines();
axs[1].gridlines();
##Longitude Labels
axs[1].set_xticks( np.arange( -180, 181, 60 ), crs = ccrs.PlateCarree() );
lon_formatter = cticker.LongitudeFormatter();
axs[1].xaxis.set_major_formatter( lon_formatter );
##Latitude labels
axs[1].set_yticks( np.arange( -90, 91, 30 ), crs = ccrs.PlateCarree() );
lat_formatter = cticker.LatitudeFormatter();
axs[1].yaxis.set_major_formatter( lat_formatter );

    
fig.subplots_adjust( bottom = 0.25, top = 0.9, left = 0.05, right = 0.95,
                   wspace = 0.1, hspace = 0.5 );
cbar_ax = fig.add_axes( [0.2, 0.2, 0.6, 0.02] );
cbar = fig.colorbar( cs, cax = cbar_ax, orientation = 'horizontal', label = 'degK' );
plt.suptitle( 'NASA GISS AMV Composite Differences' );
```
